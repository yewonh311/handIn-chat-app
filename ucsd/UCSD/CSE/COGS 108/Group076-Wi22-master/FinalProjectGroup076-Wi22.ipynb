# Setup

```python
# Imports
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
import patsy

warnings.filterwarnings("ignore")
sns.set_theme(style="ticks")
```

```python
# sets up raw_housing_data_df and raw_happiness_data_df 
raw_housing_data_df = pd.read_csv('zillow_raw_data.csv')

# This data is assumed to be from 2017 [the last known census year for this data]
raw_happiness_data_df = pd.read_csv('happiness_raw_data.csv') 

# Note source includes data from 2009-2017 might impact our time scope
bach_df = pd.read_csv("bach_by_city_raw.csv")
```

# Data Cleaning

Describe your data cleaning steps here.

- Cut all years from data except for 2017, in order to create consistency across datasets.

```python
raw_housing_data_df = raw_housing_data_df[['SizeRank', 'RegionName', 'StateName', '2017-12-31']] # '2017-12-31' == 2017 in this data

bach_df = bach_df[bach_df['year']==2017]
```

- Cleaning Functions

```python
def get_only_city_name(regionName):
    '''
        Function takes Str as input.
        Returns the last element of the String
        split on '-' character, which will be a city
        name. Return type is Str.
    '''
    return regionName.split(',')[0]

def remove_city(str_in):
    '''
        Function takes Str as input.
        Returns string missing 'city' 
        if it was present as Str.
    '''
    return str_in.replace('city', '')
```

- Clean City names across datasets, make them match eachother's formats.

```python
# Set RegionName in housing data to just be city (get rid of dashes)
raw_housing_data_df["RegionName"] = raw_housing_data_df["RegionName"].apply(get_only_city_name)

raw_housing_data_df.columns = ['City' if x=='RegionName' else x for x in raw_housing_data_df.columns]

raw_housing_data_df.columns = ['Price' if x=='2017-12-31' else x for x in raw_housing_data_df.columns]

raw_happiness_data_df["City"] = raw_happiness_data_df["City"].apply(get_only_city_name)

raw_happiness_data_df = raw_happiness_data_df[["City", "Total Score"]]
```

- Clean Bachelor's Data, add column names, set City to be consistent across dataframes.

```python
bach_df = bach_df.iloc[:,[1,3]]
bach_df.columns = ['City', 'Total Bachelors']
bach_df = bach_df.reset_index(drop=True)
bach_df[['City','State']] = bach_df.City.str.split(", ",expand=True)
bach_df['City'] = bach_df['City'].apply(remove_city)
bach_df = bach_df[['City', 'State', 'Total Bachelors']]
bach_df.columns = ['StateName' if x=='State' else x for x in bach_df.columns]
```

- Merge datasets for Analaysis

```python
total_df = raw_housing_data_df.merge(raw_happiness_data_df, how='inner', on='City')
total_df['Number Bachelors'] = bach_df["Total Bachelors"]

total_df.columns = ['Happiness Score' if x=='Total Score' else x for x in total_df.columns]

total_df = total_df.dropna()

total_df.head()
```

# Ethics & Privacy

Data bias may result from the characteristic that data does not represent the whole and only represents part of it.

First, the bias that may appear in happiness_raw_data is that since the data appear in various cities, there may be differences in income and results depending on the region. As a potential bias in datasets, data collection may have appeared differently for each city, and the method or purpose of data collection may differ for each city. In addition, even if the cities surveyed are developed countries centered on the United States, the data accuracy of other cities may be lower than that of generally industrialized cities. Because in cities that are not industrialized enough, sufficient infrastructure may not be established for data collection, and it may take more time for data to be used. So we tried to use a relatively recent dataset with relatively little difference in infrastructure. In particular, since data on education is a more sensitive factor to this, we used the Education attachment rate of 2019. Likewise, the Bachelor count by city only included data from 2017.

Another potential bias is confirmation bias. The stereotypes are that people are happy only when they achieve social achievement, and the more money they have, the less convincing even the proven objective data are. Happiness_raw_data.csv shows the total score and overall rank due to more diverse variables of emotion & physical well-being, income & employment, and community & environment, but the number of observations is low, it may be less reliable.

As data became closer to our daily lives, we all became 'data consumers'. Collecting pieces of information and data allows us to know more personal information, and if even anonymous data collects enough, he can reveal the personal information he wanted to hide and reveal who he is. However, this is when a large amount of data is sufficient, and when a pattern that the user does not even know continues to occur, but it is very unlikely that such a pattern will be found in our datasets. In addition, when looking at our data, data privacy was not a big problem in this project because it guaranteed anonymity of the people collected, who answered what results, and what situations they were in, and was investigated by reliable sources.




# Conclusion & Discussion

We conducted this research based on the motivation that we want to see the significance of a bachelor's degree on the quality of life, and we chose to quantify this with happiness and housing cost. Our initial hypothesis stood that the bachelor's degree will have a positive effect on the average happiness and housing price of the city because the it leads to a more educated, and thus "civilized" population. We gathered happiness indexes of cities from WalletHub, Housing Prices from Zillow, and baheclor's population from the trustable website diversitydatakids.org. In the EDA, we further explored our data using line graph to chart the trend of prices over different number of bachelors, as well as the heat map to discover the correlations between all the variables and note the memorable ones. 

Finally, we used Ordinary Least Square linear regression to further inspect of variables with higher (relative) correlation:
- happiness vs number of bachelors 
- price vs number of bachelors 

We noted a trivial coefficient in front of both regressor. Also, p values ( setting the signifance value at 5%), we do not reject the p value of either coefficients, thus signifying that the coefficients are not statistically significant. This leads us to **reject** our initial hypothesis and conclude that the number of bachelors in a city does not affect the housing costs or the happiness level. 

Despite these results, we were further curious to see if there was any relationship between happiness and price of housings. Using the same OLS regression (regressor = price) procedure, we found a positive coefficient which is statistically significant suggested by the p-value (p-value < 5%), which hints house price positively effects happiness.   

**Limitation**

One of the biggest limitation was that our data was not standardized. This could lead to an increased difficulty when interpreting relationships. For example the relationship between housing costs and happiness yielded a coefficient of 2.029e-05, which limits our ability to accurately analyze this information. This is because the housing costs is siginificatly larger than the numerical value of happiness index.

Secondly, in data wrangling, we only kept cities whose number of bachelors, housing costs, and happiness index we were able to find. This causes the sample selection to not be random and detriments the accuracy of OLS regression especially when correlating cities with lower population (thus availibility of data).

Finally, the inherent subjective nature of happiness indicates that any the correlationship drawn using the happiness index might be trivial in the context and interpretation of each city.
